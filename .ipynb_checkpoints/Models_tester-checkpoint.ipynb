# Models Test
This is a notebook that you can try different pretrained models, specify the dir to pretrained models and content images as the next cell shows.

```python
import torch
from torchvision import transforms
#from torch.optim import Adam
from torchvision import datasets
from torch.utils.data import DataLoader

#import time

from helpers import *
from style_subnet import Style_Subnet
from enhance_subnet import Enhance_Subnet
from refine_subnet import Refine_Subnet
#from vgg import Vgg19, Vgg16

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# path to pretrained model
style_dir = 'pretrained_models/henri'
# path to content images, make sure the images inside a sub folder of the current one
# why? pytorch datasets.ImageFolder() asks so...
content_dir = 'stylized_imgs/0_Original/'
```

```python
content_trans = transforms.Compose([#transforms.Resize(480),
                                    #transforms.CenterCrop(480),
                                    transforms.Resize(256),
                                    transforms.ToTensor(),
                                    tensor_normalizer()])

train_dataset = datasets.ImageFolder(root=content_dir, transform=content_trans)
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True)

# define 3 subnetwork and load the pretrained weights
style_subnet = Style_Subnet().to(device)
enhance_subnet = Enhance_Subnet().to(device)
refine_subnet = Refine_Subnet().to(device)

style_subnet.load_state_dict(torch.load(style_dir+'_St.pt'))
enhance_subnet.load_state_dict(torch.load(style_dir+'_En.pt'))
refine_subnet.load_state_dict(torch.load(style_dir+'_Re.pt'))
```

```python
x, _ = next(iter(train_loader))
imshow(x)
```

```python
y1 = style_subnet(x.to(device))
print(y1[0].max())
print(y1[0].min())
y2 = enhance_subnet(y1[0])
print(y2[0].max())
print(y2[0].min())
y3 = refine_subnet(y2[0])
print(y3[0].max())
print(y3[0].min())
```
